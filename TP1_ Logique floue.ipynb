```python

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Univers de discours
x_distance = np.linspace(0, 30, 300)
```

```python
# DÃ©finition des ensembles flous
v_small = fuzz.trapmf(x_distance, [0, 0, 3, 6])
small   = fuzz.trimf(x_distance, [4, 8, 12])
perfect = fuzz.trimf(x_distance, [10, 15, 20])
big     = fuzz.trimf(x_distance, [18, 22, 26])
v_big   = fuzz.trapmf(x_distance, [24, 27, 30, 30])
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x_distance, v_small, 'b', label='v_small')
plt.plot(x_distance, small, 'c', label='small')
plt.plot(x_distance, perfect, 'g', label='perfect')
plt.plot(x_distance, big, 'orange', label='big')
plt.plot(x_distance, v_big, 'r', label='v_big')

plt.title("Fonctions d'appartenance de la variable Distance_to_next_car")
plt.xlabel("Distance (mÃ¨tres)")
plt.ylabel("DegrÃ© d'appartenance")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
# Importation des bibliothÃ¨ques nÃ©cessaires
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# 1 DÃ©finir lâunivers de discours pour Speed_change
speed_change = np.arange(-20, 20.1, 0.1)

# 2 DÃ©finir les ensembles flous (fonctions dâappartenance)
declining = fuzz.trapmf(speed_change, [-20, -20, -10, -2])   # vitesse diminue
constant  = fuzz.trimf(speed_change, [-5, 0, 5])             # vitesse stable
growing   = fuzz.trapmf(speed_change, [2, 10, 20, 20])       # vitesse augmente

# 3  Tracer les fonctions dâappartenance
plt.figure(figsize=(10,6))
plt.plot(speed_change, declining, 'b', linewidth=2, label='declining')
plt.plot(speed_change, constant, 'g', linewidth=2, label='constant')
plt.plot(speed_change, growing, 'r', linewidth=2, label='growing')

plt.title('Fonctions dâappartenance de la variable floue "Speed_change"')
plt.xlabel('Variation de vitesse (km/h ou m/s)')
plt.ylabel('DegrÃ© dâappartenance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importation des bibliothÃ¨ques nÃ©cessaires
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# 1 DÃ©finir lâunivers de discours pour Acceleration_Adjust
acc_adjust = np.arange(-3, 3.1, 0.1)

# 2 DÃ©finir les ensembles flous (fonctions dâappartenance)
neg_big   = fuzz.trapmf(acc_adjust, [-3, -3, -2.5, -1.5])   # forte dÃ©cÃ©lÃ©ration
neg_small = fuzz.trimf(acc_adjust, [-2, -1, 0])             # faible dÃ©cÃ©lÃ©ration
zero      = fuzz.trimf(acc_adjust, [-0.5, 0, 0.5])          # aucune variation
pos_small = fuzz.trimf(acc_adjust, [0, 1, 2])               # faible accÃ©lÃ©ration
pos_big   = fuzz.trapmf(acc_adjust, [1.5, 2.5, 3, 3])       # forte accÃ©lÃ©ration

# 3 Tracer les fonctions dâappartenance
plt.figure(figsize=(10,6))
plt.plot(acc_adjust, neg_big, 'r', linewidth=2, label='neg_big')
plt.plot(acc_adjust, neg_small, 'orange', linewidth=2, label='neg_small')
plt.plot(acc_adjust, zero, 'g', linewidth=2, label='zero')
plt.plot(acc_adjust, pos_small, 'b', linewidth=2, label='pos_small')
plt.plot(acc_adjust, pos_big, 'm', linewidth=2, label='pos_big')

plt.title('Fonctions dâappartenance de la variable floue "Acceleration_Adjust"')
plt.xlabel('Ajustement de lâaccÃ©lÃ©ration (m/sÂ²)')
plt.ylabel('DegrÃ© dâappartenance')
plt.legend()
plt.grid(True)
plt.show()
```

## EXERCICE 4

```python
# Importation des bibliothÃ¨ques nÃ©cessaires
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# ==========================================================
# 1 DÃ©finition des variables floues
# ==========================================================

# Variables d'entrÃ©e
Distance_to_next_car = ctrl.Antecedent(np.arange(0, 31, 1), 'Distance_to_next_car')
Speed_change = ctrl.Antecedent(np.arange(-20, 21, 1), 'Speed_change')

# Variable de sortie
Acceleration_Adjust = ctrl.Consequent(np.arange(-3, 3.1, 0.1), 'Acceleration_Adjust')

# ==========================================================
# 2 Fonctions dâappartenance
# ==========================================================

# --- Distance_to_next_car ---
Distance_to_next_car['v_small'] = fuzz.trapmf(Distance_to_next_car.universe, [0, 0, 2, 6])
Distance_to_next_car['small']   = fuzz.trimf(Distance_to_next_car.universe, [4, 8, 12])
Distance_to_next_car['perfect'] = fuzz.trimf(Distance_to_next_car.universe, [10, 15, 20])
Distance_to_next_car['big']     = fuzz.trimf(Distance_to_next_car.universe, [18, 22, 26])
Distance_to_next_car['v_big']   = fuzz.trapmf(Distance_to_next_car.universe, [24, 28, 30, 30])

# --- Speed_change ---
Speed_change['declining'] = fuzz.trapmf(Speed_change.universe, [-20, -20, -10, -2])
Speed_change['constant']  = fuzz.trimf(Speed_change.universe, [-3, 0, 3])
Speed_change['growing']   = fuzz.trapmf(Speed_change.universe, [2, 10, 20, 20])

# --- Acceleration_Adjust ---
Acceleration_Adjust['neg_big']   = fuzz.trapmf(Acceleration_Adjust.universe, [-3, -3, -2.5, -1.5])
Acceleration_Adjust['neg_small'] = fuzz.trimf(Acceleration_Adjust.universe, [-2, -1, 0])
Acceleration_Adjust['zero']      = fuzz.trimf(Acceleration_Adjust.universe, [-0.5, 0, 0.5])
Acceleration_Adjust['pos_small'] = fuzz.trimf(Acceleration_Adjust.universe, [0, 1, 2])
Acceleration_Adjust['pos_big']   = fuzz.trapmf(Acceleration_Adjust.universe, [1.5, 2.5, 3, 3])

# ==========================================================
# 3 Visualisation des fonctions dâappartenance
# ==========================================================
Distance_to_next_car.view()
Speed_change.view()
Acceleration_Adjust.view()

# ==========================================================
# 4 DÃ©finition des rÃ¨gles floues (ctrl.Rule)
# ==========================================================

rules = [
    ctrl.Rule(Distance_to_next_car['v_small'] & Speed_change['declining'], Acceleration_Adjust['neg_small']),
    ctrl.Rule(Distance_to_next_car['v_small'] & Speed_change['constant'],  Acceleration_Adjust['neg_big']),
    ctrl.Rule(Distance_to_next_car['v_small'] & Speed_change['growing'],   Acceleration_Adjust['neg_big']),

    ctrl.Rule(Distance_to_next_car['small'] & Speed_change['declining'], Acceleration_Adjust['zero']),
    ctrl.Rule(Distance_to_next_car['small'] & Speed_change['constant'],  Acceleration_Adjust['neg_small']),
    ctrl.Rule(Distance_to_next_car['small'] & Speed_change['growing'],   Acceleration_Adjust['neg_big']),

    ctrl.Rule(Distance_to_next_car['perfect'] & Speed_change['declining'], Acceleration_Adjust['pos_small']),
    ctrl.Rule(Distance_to_next_car['perfect'] & Speed_change['constant'],  Acceleration_Adjust['zero']),
    ctrl.Rule(Distance_to_next_car['perfect'] & Speed_change['growing'],   Acceleration_Adjust['neg_small']),

    ctrl.Rule(Distance_to_next_car['big'] & Speed_change['declining'], Acceleration_Adjust['pos_big']),
    ctrl.Rule(Distance_to_next_car['big'] & Speed_change['constant'],  Acceleration_Adjust['pos_small']),
    ctrl.Rule(Distance_to_next_car['big'] & Speed_change['growing'],   Acceleration_Adjust['zero']),

    ctrl.Rule(Distance_to_next_car['v_big'] & Speed_change['declining'], Acceleration_Adjust['pos_big']),
    ctrl.Rule(Distance_to_next_car['v_big'] & Speed_change['constant'],  Acceleration_Adjust['pos_big']),
    ctrl.Rule(Distance_to_next_car['v_big'] & Speed_change['growing'],   Acceleration_Adjust['pos_small'])
]

# ==========================================================
# 5 CrÃ©ation du systÃ¨me de contrÃ´le flou
# ==========================================================
speed_ctrl = ctrl.ControlSystem(rules)
speed_sim = ctrl.ControlSystemSimulation(speed_ctrl)

# ==========================================================
# 6 Simulation pour Distance=13 et Speed_change=-2.5
# ==========================================================
speed_sim.input['Distance_to_next_car'] = 13
speed_sim.input['Speed_change'] = -2.5

speed_sim.compute()

print("Valeur dÃ©fuzzifiÃ©e de Acceleration_Adjust : {:.3f}".format(speed_sim.output['Acceleration_Adjust']))

# Visualisation du rÃ©sultat
Acceleration_Adjust.view(sim=speed_sim)
plt.show()
```
